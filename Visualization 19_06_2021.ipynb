## Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(color_codes = True)
```

```python
pd.set_option('display.max_columns',None)
```

```python
auto  = pd.read_csv("Automobile.csv")
```

```python
auto.head()
```

```python
auto['symboling'].unique()
```

```python
auto.shape
```

```python
sns.histplot(data = auto,x = 'normalized_losses',color = "red",bins = 20,hue = 'fuel_type')
```

```python
sns.jointplot(auto['fuel_type'],auto['horsepower'])
```

```python
sns.countplot(auto['fuel_type'])
```

```python
auto['fuel_type'].value_counts()
```

```python
sns.jointplot(data = auto,x = 'horsepower', y = 'price',kind = "reg" )
plt.title("Horse Power vs Price")
```

```python
sns.jointplot(data = auto,x = 'horsepower', y = 'highway_mpg',kind = "reg" )
```

```python
auto.columns
```

```python
sns.pairplot(auto[['normalized_losses','engine_size','compression_ratio','horsepower','price']])
```

```python
sns.boxplot(data = auto ,x = 'number_of_doors',y = 'horsepower')
```

```python
sns.countplot(auto['body_style'])
```

```python
auto.dtypes
```

```python
sns.lmplot(data = auto,x = "horsepower",y = "price",hue = 'fuel_type')
```
